## Introduction

Les amplificateurs de classe A sont les amplificateur linéaire les plus fidèles. Cela veut dire que la distorsion du signal atttendu doit être très faible. Néanmoins les rendements sont très faible. Donc leur usage est généralement limité aux amplificateurs de très faible puissance ou encore aux amplificateurs haute-fidélité haut de gamme de puissance moyenne.

Ce type d'amplificateur se construit avec un amplificateur de sortie. Le point se trouve au milieux de la droite de charge.
